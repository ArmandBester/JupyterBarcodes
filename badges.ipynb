```python
xOrg = 20
width = 98

def make_label_2(xOrg=xOrg, yOrg=20, width = width, height = 120, name="Armand Bester", aff="University of the Free State", country="South Africa"):   
    pdf.set_fill_color(255)
    pdf.set_draw_color(r=125, g=125, b=125)
    pdf.rect(xOrg, y=yOrg, w=width, h=height, style="FD")
    
    
    txOrg = xOrg + 10
    tyOrg = yOrg + 20

    
    #name
    pdf.set_font(family="Arial", style='', size=20)
    pdf.text(x = txOrg, y= tyOrg, txt="Name:")
    pdf.text(x = txOrg+10, y= tyOrg+10, txt=name)
    #Affiliation
    pdf.text(x = txOrg, y= tyOrg+30, txt="Affiliation:")
    pdf.set_font(family="Arial", style='', size=15)
    pdf.text(x = txOrg+10, y= tyOrg+40, txt=aff)
    #Country
    pdf.set_font(family="Arial", style='', size=20)
    pdf.text(x = txOrg, y= tyOrg+75, txt="Country:")
    pdf.text(x = txOrg+10, y= tyOrg+85, txt=country)


modelDict = {0: xOrg, 
            1: xOrg+width}

pdf = FPDF(orientation="landscape", format="A4")
for p in range(10):
    #create a new page
    pdf.add_page()      

    #add the two rectangels on the new page
    for i in range(0,2):
        model = i%2
        if model == 0:
            make_label_2(modelDict[model])
        elif model == 1:
            make_label_2(modelDict[model])
            
pdf.output("squares.pdf")
```
